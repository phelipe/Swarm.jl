A seguinte função possui mínimo em [1,1] com o valor de 4

```julia
function myfunc(x::Vector)
    x1, x2 = x[1], x[2]
    return x1^4 - 2*x2*x1^2 + x2^2 + x1^2 - 2*x1 + 5
end

myfunc(x1, x2) = myfunc([x1, x2])
    
lb = [-3.0, -1.0]
ub = [2.0, 6.0]
x, y = -3.:0.1:2, -1.:0.1:6;
#plot(x,y,myfunc, st = [:surface])
```

```julia
    #"  Yield Stress <= 100 kpsi")
    #"  Yield Stress <= Buckling Stress")
    #"  Deflection   <= 0.25 inches")

    function weight(x, B, ρ, E, P)
        H, d, t = x  # all in inches
        return ρ*2*π*d*t*√((B/2)^2 + H^2)
    end

    function stress(x, B, ρ, E, P)
        H, d, t = x  # all in inches
        return (P * √((B/2)^2 + H^2)) / (2*t*π*d*H)
    end

    function buckling_stress(x, B, ρ, E, P)
        H, d, t = x  # all in inches
        return (π^2 * E * (d^2 + t^2)) / (8 * ((B/2)^2 + H^2))
    end

    function deflection(x, B, ρ, E, P)
        H, d, t = x  # all in inches
        return (P * (√((B/2)^2 + H^2))^3) / (2*t*π*d*H^2 * E)
    end

    function mycons(x, B, ρ, E, P)
        strs = stress(x, B, ρ, E, P)
        buck = buckling_stress(x, B, ρ, E, P)
        defl = deflection(x, B, ρ, E, P)
        return [100 - strs, buck - strs, 0.25 - defl]
    end

    B = 60  # inches
    rho = 0.3  # lb/in^3
    E = 30000  # kpsi
    P = 66  # lb (force)
    args = (B, rho, E, P)
    lb = [10.0, 1.0, 0.01]
    ub = [30.0, 3.0, 0.25];
```

```julia
#pso(weight, lb, ub, constraints=mycons, args=args, swarmsize=1000)
particle = ihpso(weight, lb, ub, verbose= false, constraints=mycons, args=args,
    maxiter = 500, swarmsize = 300, minfunc = 1e-6);
println("  Yield Stress <= 100 kpsi")
println("  Yield Stress <= Buckling Stress")
println("  Deflection   <= 0.25 inches")
println("----------------------------------------------")
println("The optimum is at:")
println("    $(particle.best_position)")
println("Optimal function values:")
println("    weight         : $(particle.best_value)")
println("Constraint functions:")
println("    yield stress   : $(stress(particle.best_position, args...))")
println("    buckling stress: $(buckling_stress(particle.best_position, args...))")
println("    deflection     : $(deflection(particle.best_position, args...))")
plot(particle.history)
```

```julia
#pso(weight, lb, ub, constraints=mycons, args=args, swarmsize=1000)
particle = hpso(weight, lb, ub, verbose= false, constraints=mycons, args=args,
    maxiter = 500, swarmsize = 300, minfunc = 1e-6);
println("  Yield Stress <= 100 kpsi")
println("  Yield Stress <= Buckling Stress")
println("  Deflection   <= 0.25 inches")
println("----------------------------------------------")
println("The optimum is at:")
println("    $(particle.best_position)")
println("Optimal function values:")
println("    weight         : $(particle.best_value)")
println("Constraint functions:")
println("    yield stress   : $(stress(particle.best_position, args...))")
println("    buckling stress: $(buckling_stress(particle.best_position, args...))")
println("    deflection     : $(deflection(particle.best_position, args...))")
plot(particle.history)
```
